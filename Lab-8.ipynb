```python
#Question-3

class RealNumber:
    def __init__(self, r=0):
        self.__realValue = r
    def getRealValue(self):
        return self.__realValue
    def setRealValue(self, r):
        self.__realValue = r
    def ping(self):
        print('I am in RealNumber class')
    def __str__(self):
        return 'RealPart: '+str(self.getRealValue())
    
###############################
class ComplexNumber(RealNumber):
    def __init__(self, r=1,c=1):
        super().__init__(r)
        self.com = c
    def __str__(self):
        return f"RealPart: {self.getRealValue()}.0\nImaginaryPart: {(self.com)}.0"
#####################################    
    
    
cn1 = ComplexNumber()
print(cn1)
print('---------')
cn2 = ComplexNumber(5,7)
print(cn2)
print('---------')
```

```python
#Question-5

class Animal:
    def __init__(self,sound):
        self.__sound = sound

    def makeSound(self):
        return self.__sound
    
class Printer:
    def printSound(self, a):
        print(a.makeSound())
        
#########################
class Dog(Animal):
    pass
     
class Cat(Animal):
    pass
######################    

d1 = Dog('bark')
c1 = Cat('meow')
a1 = Animal('Animal does not make sound')
pr = Printer()
pr.printSound(a1)
pr.printSound(c1)
pr.printSound(d1)
```

```python
#Quesion - 6

class Fruit:
    def __init__(self, formalin=False, name=''):
        self.__formalin = formalin
        self.name = name
 
    def getName(self):
        return self.name

    def hasFormalin(self):
        return self.__formalin

class testFruit:
    def test(self, f):
        print('----Printing Detail----')
        if f.hasFormalin():
            print('Do not eat the ',f.getName(),'.')
            print(f)
        else:
            print('Eat the ',f.getName(),'.')
            print(f)
########################
class Mango(Fruit):
    def __init__(self,name="Mango"):
        super().__init__(name)
    def __str__(self):
        return "Mangos are bad for you"
    
class Jackfruit(Fruit):
    def __init__(self,formalin=True,name="Jackfruit"): 
        super().__init__(formalin,name)
    def __str__(self):
        return "Jackfruits are good for you"

############################
m = Mango()
j = Jackfruit()
t1 = testFruit()
t1.test(m)
t1.test(j)


```
